# Gráficos

```python
plt.style.use("ggplot")
# converte o mapa de cores em um vetor, para uso de cores individuais
vcmap = plt.get_cmap("viridis", 6)
coresViridis = [mpl.colors.rgb2hex(vcmap(i)) for i in range(6)] # extrai lista de cores em hex
display(coresViridis)
```

```python
plt.plot(df.Brasil)
```

```python
savefig,fig=plt.subplots(figsize=(25,10))

# fig.plot(df.Brasil/1000000,color=coresViridis[0],label='Brasil',marker='o')
fig.plot(df.Nordeste/1000000,color=coresViridis[1],label='Nordeste',marker='o')
fig.plot(df.Sudeste/1000000,color=coresViridis[2],label='Sudeste',marker='o')
fig.plot(df.Sul/1000000,color=coresViridis[3],label='Sul',marker='o')
fig.plot(df['Centro-Oeste']/1000000,color=coresViridis[4],label='Centro-Oeste',marker='o')

fig.set_xticks(ticks = df.index, labels = df.index)

fig.legend(fontsize=15)

handles = fig.get_legend_handles_labels()
texts=[]
for i in range(len(handles[0])):
  endx = handles[0][i].get_xdata()[-1] # dados X da linha
  endy = handles[0][i].get_ydata()[-1] # dados Y da linha
  label = handles[1][i]                # rótulo da linha (parâmetro label no Axes.plot)
  #nomes das séries ao final de cada linha
  #if/else remove o 'Centro-Oeste' do ajuste (at.adjust.text)
  if i<3:
    texts.append(fig.annotate(text = label, xy = (endx+1, endy-3),
                              color = handles[0][i].get_color(), size = 15))
  else:
    fig.annotate(text = label, xy = (endx+1, endy-3),
                 color = handles[0][i].get_color(), size = 15)
  # valor em cada ponto das linhas
  for j in range(len(handles[0][i].get_xdata())):
    x=handles[0][i].get_xdata()[j]
    y=handles[0][i].get_ydata()[j]
    texts.append(fig.annotate(text='{:.2f}'.format(y),xy=(x,y),size=10,
                              color=handles[0][i].get_color(),
                              path_effects=[pe.withStroke(linewidth=3,
                                                          foreground="black")]))

plt.xlim([min(df.index), max(df.index) + 10 ])

plt.title('Evolução da população brasileira por região')
plt.ylabel('população (em milhões de habitantes)')

at.adjust_text(texts)

# plt.savefig("abc.svg")
plt.show()
```
